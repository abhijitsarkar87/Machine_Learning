# Linear Regression

###  1. Formula :

         a. Y= B0+B1xi+Error # simple Linear Regression
         b. Y=B0+B1x1+B2x2..........Bnxn+Error # Multiple Linear Regression 
        
         Y = Outcome
         B0= Intercept -----> From where imagnary line is drawn
         B1= coefficient of x
        
        AND........... x=is the independent variable................
        
        ..
        
        
 ### 2. Error:
              
              Sum of (y-y[hat])**2
              
              1. y= Original
              2. y[hat]= Predicted
