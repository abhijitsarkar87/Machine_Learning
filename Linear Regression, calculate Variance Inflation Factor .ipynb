# Linear Regression

###  1. Formula :

         a. Y= B0+B1xi+Error # simple Linear Regression
         b. Y=B0+B1x1+B2x2..........Bnxn+Error # Multiple Linear Regression 
        
         Y = Outcome
         B0= Intercept -----> From where imagnary line is drawn
         B1= coefficient of x
        
        AND........... x=is the independent variable................
        
 ### 2. Error:
              
              Sum of (y-y[hat])**2
              
              1. y= Original
              2. y[hat]= Predicted

```python
bos_new.head()
```

```python
x=bos_new.iloc[:,1:-1]
x.sample()
```

```python
y=bos_new.iloc[:,[-1]]
y.head()
```

# Process Linear Regression..............................

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)

# checking shape
print("Train_x: ",x_train.shape,"Test_x: ",x_test.shape,"\nTrain_y: ",y_train.shape,"Test_y",y_test.shape)
```

```python
# create instance
from sklearn.linear_model import LinearRegression
lm=LinearRegression()

# training (train model) 
lm.fit(x_train,y_train)

#coefficient or slope value
print("Cofficient_ Values:\n",lm.coef_)  # if +ve would postive that would add to house price, if (-ve) it would add effect the house price

# intercept (B0 or a)
print("\n\nIntercept_Values: ",lm.intercept_) 

# score
print("\n\nTraining Score: ",lm.score(x_train,y_train))
```

# Testing

```python
pred=lm.predict(x_test)
print("Prediced Result\n",pred)
print("Actual Result\n",y_test)
```

# Error

```python
#**NOTE: Pred is the instance for machine given answer 
# y_test is the actual answers, 

# Error_Formula: Sum of (Actual-Machine_Predicted Result)**2

print("Mean Absolute Error :",mean_absolute_error(y_test,pred))

print("Mean Squared Error  :",mean_squared_error(y_test,pred))

print("Root Mean Squared Error :",np.sqrt(mean_squared_error(y_test,pred))) # square root of Error Formula
```

## R2 Score

```python
# r2-score: Coefficient of determination is the propertional of the variance in the dependent variable that is predictable from the independent varible
# i.e. Change coming in y whenever x is being changed
# call R2_Score>> from Metrics 

from sklearn.metrics import r2_score
```

## calculate Variance Inflasion Factor

```python
#import statsmodel.api as sm
import pandas as pd
import scipy.stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(dataset_name):
    vif=pd.DataFrame()# create data frame named vif
    # add columns
    vif['Variables']=dataset_name.columns # insert column names in variable columns
    vif['VIF FACTOR']=[variance_inflation_factor(dataset_name.values,i) for i in range(dataset_name.shape[1])]# for i range(column_size) 
    return vif
```
